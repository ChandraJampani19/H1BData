```python
from google.cloud import bigquery
import pandas as pd


client = bigquery.Client(project='datenglea')

# Define the dataset and table name
dataset_id = 'H1BData'
table_id = 'H1practisetable'  # Replace with your desired table name

# Define the schema for the table
schema = [
    bigquery.SchemaField('Record', 'INT64'),  # Rename 'Unnamed: 0' to 'id'
    bigquery.SchemaField('CASE_STATUS', 'STRING'),
    bigquery.SchemaField('EMPLOYER_NAME', 'STRING'),
    bigquery.SchemaField('SOC_NAME', 'STRING'),
    bigquery.SchemaField('JOB_TITLE', 'STRING'),
    bigquery.SchemaField('FULL_TIME_POSITION', 'STRING'),
    bigquery.SchemaField('PREVAILING_WAGE', 'FLOAT64'),  # Assuming numeric values
    bigquery.SchemaField('YEAR', 'INT64'),
    bigquery.SchemaField('WORKSITE', 'STRING'),
    bigquery.SchemaField('lon', 'FLOAT64'),
    bigquery.SchemaField('lat', 'FLOAT64'),
]


# Define the table reference
table_ref = client.dataset(dataset_id).table(table_id)

# Create the table
table = bigquery.Table(table_ref, schema=schema)
table = client.create_table(table)

print(f'Table {table_id} created in dataset {dataset_id}.')
```

```python
from google.cloud import bigquery

# Initialize BigQuery client
client = bigquery.Client(project='datenglea')

# Define the table reference
dataset_id = 'H1BData'
table_id = 'H1practisetable'
table_ref = client.dataset(dataset_id).table(table_id)

# Run a query to count the number of rows in the table
query = f"SELECT COUNT(*) as record_count FROM `{dataset_id}.{table_id}`"
query_job = client.query(query)

display = f"SELECT *  FROM `{dataset_id}.{table_id}`LIMIT 10"
d_query_job = client.query(display)

for row in d_query_job.result():
    print(row)
# Get the results
results = query_job.result()

# Print the count
for row in results:
    print(f"Total records: {row.record_count}")
```
